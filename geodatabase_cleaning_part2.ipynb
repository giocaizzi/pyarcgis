```python
import pandas as pd
import json
```

```python
with open('tavole.json',"r") as f:
    tavole = json.load(f)   
```

```python
tavole
```

```python
rows = []
for upper_key, inner_dict in tavole.items():
    for lower_key, value in inner_dict.items():
        if isinstance(value, dict):
            for nested_key, nested_value in value.items():
                rows.append((upper_key, lower_key, nested_key, nested_value))
        elif isinstance(value, list):
            for i, list_value in enumerate(value):
                rows.append((upper_key, lower_key, i+1, list_value))
        else:
            raise TypeError("Unexpected type")
columns=['Filename', 'LayerType', 'Layer name', 'DataSource path']
df = pd.DataFrame(rows, columns=columns)

# Set the index
df.set_index(['Filename', 'LayerType', 'Layer name',], inplace=True)
df.sort_index(inplace=True)
```
